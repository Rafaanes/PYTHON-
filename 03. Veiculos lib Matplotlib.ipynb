```python
# Importa a biblioteca pandas com o apelido "pd"  usada para trabalhar com tabelas de dados
import pandas as pd

# Importa matplotlib.pyplot com o apelido "plt"  usada para criar grficos
import matplotlib.pyplot as plt

# Caminho do arquivo CSV contendo os dados dos veculos
# Substitua o caminho pelo local onde o arquivo est salvo
# 02_Veiculos.csv
arquivo_csv = r'C:\Rafalla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\02_Veiculos.csv'

# L o arquivo CSV e carrega os dados em um DataFrame chamado "df"
df = pd.read_csv(arquivo_csv, sep= ',', encoding = 'latin-1')

# Seleciona apenas as colunas que vamos usar e remove linhas com dados ausentes
df = df[['Engine Fuel Type', 'MSRP']].dropna()

# Agrupa os dados por tipo de combustvel
agrupado = df.groupby('Engine Fuel Type')

# Calcula o preo mdio (MSRP) para cada tipo de combustvel
media_preco = agrupado['MSRP'].mean()

# Conta quantos veculos existem de cada tipo de combustvel
frequencia = agrupado.size()

# Cria uma nova figura de tamanho 12x6 polegadas e um conjunto de eixos (ax1)
#subplots = grfico de coluna e linhas
fig, ax1 = plt.subplots(figsize = (12,6))

# ---------- GRFICO DE BARRAS ----------
# Plota as barras com os preos mdios
# Argumentos:
# - media_preco.index  rtulos no eixo X (tipos de combustvel)
# - media_preco.values  altura das barras (valores mdios)
# - color='green'  COR DAS BARRAS (Voc pode trocar aqui, exemplo: 'red', 'orange', '#FF5733', etc.)
# - label  legenda (no obrigatria, mas til se usar .legend())
ax1.bar(media_preco.index,media_preco.values, color = 'skyblue', label = 'Preo Medio (MSRP)')

# Define o nome do eixo X
ax1.set_xlabel('Tipo de Combustvel')

# Define o nome do eixo Y (lado esquerdo), referente ao preo
ax1.set_ylabel ('Preo Mdio (MSRP)', color = 'green')

# Define a cor dos nmeros do eixo Y esquerdo (para combinar com as barras)
ax1.tick_params(axis='y', labelcolor = 'green')

# Gira os nomes no eixo X para facilitar a leitura
ax1.set_xticklabels(media_preco.index, rotation = 45, ha='right')

# ---------- SEGUNDO EIXO Y (DIREITO) ----------
# Cria um segundo eixo Y (compartilhando o mesmo eixo X)
ax2 = ax1.twinx()

# Plota a linha de quantidade de veculos por tipo
# Argumentos:
# - media_preco.index  rtulos no eixo X
# - frequencia.values  valores para a linha
# - color  cor da linha
# - marker='o'  adiciona crculos nos pontos da linha
ax2.plot(media_preco.index, frequencia.values, color = 'green', marker = '*', label = 'Quantidade de Veculos')

# Define o nome do eixo Y (lado direito), referente  quantidade
ax2.set_ylabel('Quantidade de Veculos', color = 'blue')

# Define a cor dos nmeros do eixo Y direito
ax2.tick_params(axis='y', labelcolor = 'blue')

# Define o ttulo geral do grfico
plt.title('Mdia de Preo (MSRP) e Quantidade de Veculos por tipo de Combustvel')

# Ajusta automaticamente o layout para que nada fique cortado
fig.tight_layout()

# Ativa a grade no fundo do grfico (linhas horizontais)
# plt.grid(True)

# Exibe o grfico na tela
plt.show()




```
