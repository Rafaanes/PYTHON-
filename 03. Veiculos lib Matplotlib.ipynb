```python
# Importa a biblioteca pandas com o apelido "pd" — usada para trabalhar com tabelas de dados
import pandas as pd

# Importa matplotlib.pyplot com o apelido "plt" — usada para criar gráficos
import matplotlib.pyplot as plt

# Caminho do arquivo CSV contendo os dados dos veículos
# Substitua o caminho pelo local onde o arquivo está salvo
# 02_Veiculos.csv
arquivo_csv = r'C:\Rafaélla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\02_Veiculos.csv'

# Lê o arquivo CSV e carrega os dados em um DataFrame chamado "df"
df = pd.read_csv(arquivo_csv, sep= ',', encoding = 'latin-1')

# Seleciona apenas as colunas que vamos usar e remove linhas com dados ausentes
df = df[['Engine Fuel Type', 'MSRP']].dropna()

# Agrupa os dados por tipo de combustível
agrupado = df.groupby('Engine Fuel Type')

# Calcula o preço médio (MSRP) para cada tipo de combustível
media_preco = agrupado['MSRP'].mean()

# Conta quantos veículos existem de cada tipo de combustível
frequencia = agrupado.size()

# Cria uma nova figura de tamanho 12x6 polegadas e um conjunto de eixos (ax1)
#subplots = gráfico de coluna e linhas
fig, ax1 = plt.subplots(figsize = (12,6))

# ---------- GRÁFICO DE BARRAS ----------
# Plota as barras com os preços médios
# Argumentos:
# - media_preco.index → rótulos no eixo X (tipos de combustível)
# - media_preco.values → altura das barras (valores médios)
# - color='green' → COR DAS BARRAS (Você pode trocar aqui, exemplo: 'red', 'orange', '#FF5733', etc.)
# - label → legenda (não obrigatória, mas útil se usar .legend())
ax1.bar(media_preco.index,media_preco.values, color = 'skyblue', label = 'Preço Medio (MSRP)')

# Define o nome do eixo X
ax1.set_xlabel('Tipo de Combustível')

# Define o nome do eixo Y (lado esquerdo), referente ao preço
ax1.set_ylabel ('Preço Médio (MSRP)', color = 'green')

# Define a cor dos números do eixo Y esquerdo (para combinar com as barras)
ax1.tick_params(axis='y', labelcolor = 'green')

# Gira os nomes no eixo X para facilitar a leitura
ax1.set_xticklabels(media_preco.index, rotation = 45, ha='right')

# ---------- SEGUNDO EIXO Y (DIREITO) ----------
# Cria um segundo eixo Y (compartilhando o mesmo eixo X)
ax2 = ax1.twinx()

# Plota a linha de quantidade de veículos por tipo
# Argumentos:
# - media_preco.index → rótulos no eixo X
# - frequencia.values → valores para a linha
# - color → cor da linha
# - marker='o' → adiciona círculos nos pontos da linha
ax2.plot(media_preco.index, frequencia.values, color = 'green', marker = '*', label = 'Quantidade de Veículos')

# Define o nome do eixo Y (lado direito), referente à quantidade
ax2.set_ylabel('Quantidade de Veículos', color = 'blue')

# Define a cor dos números do eixo Y direito
ax2.tick_params(axis='y', labelcolor = 'blue')

# Define o título geral do gráfico
plt.title('Média de Preço (MSRP) e Quantidade de Veículos por tipo de Combustível')

# Ajusta automaticamente o layout para que nada fique cortado
fig.tight_layout()

# Ativa a grade no fundo do gráfico (linhas horizontais)
# plt.grid(True)

# Exibe o gráfico na tela
plt.show()




```
