```python
'''

Enunciado do Projeto: Análise de Vendas

Contexto:
========

Você foi contratado por uma pequena empresa fictícia chamada 
'Comercial Silva' para realizar uma análise simples das vendas 
mensais do último ano. O objetivo é obter estatísticas 
básicas e visualizar o desempenho mensal para apoiar 
futuras decisões da empresa.


O que você deve fazer:
======================

1)  Criar um arquivo Excel contendo os dados de vendas da empresa mês a mês.

2)  Utilizar as bibliotecas pandas, numpy e matplotlib para:

        >> Ler os dados do arquivo Excel.

        >> Calcular estatísticas básicas:

            Soma total de vendas no ano.
            Média mensal.
            Mediana.
            Desvio padrão.

        >> Exibir os dados no terminal.

3)  Criar um gráfico de linha para mostrar as vendas 
    ao longo dos 12 meses.

    


Objetivo final:
==============

    Ter um arquivo Excel com os dados.

    Um programa Python funcional que leia esse arquivo e gere:

    Estatísticas simples.


    Um gráfico claro e bonito para apresentação visual.   

    Biblioteca	Função no código
    ==========  ================
    pandas	    Manipular e salvar os dados da planilha Excel
    numpy	    Calcular média, mediana, soma, desvio padrão etc.
    matplotlib	Criar o gráfico de vendas
    openpyxl	Ler/gravar arquivos .xlsx no Excel com o pandas

    ATENÇÃO: 
    =======
    Abra o terminal e instale:

    pip install pandas 
    pip install numpy 
    pip install matplotlib 
    pip install openpyxl

'''



# Importa as bibliotecas necessárias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


# 1. Criar o arquivo Excel com os dados da empresa


# Criamos os dados de vendas da empresa mês a mês
#{} = chaves significa que temos uma coleção de dados e dentro dela temos duas lista.
dados = {
    'Mês': ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho',
            'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro'],
    'Vendas (R$)': [15000, 18000, 12000, 19500, 21000, 17000,
                    16000, 22000, 20000, 23000, 25000, 30000]
}


# Transforma os dados em uma tabela (DataFrame)
df = pd.DataFrame(dados)


# Define o nome do arquivo Excel que será criado
# 01_vendas_empresa.xlsx
arquivo_excel = r'C:\Rafaélla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Projeto\Bases Projeto\01.vendas empresa.xlsx'


# Salva a tabela no arquivo Excel
df.to_excel(arquivo_excel, index=False)

# Informa ao usuário que o arquivo foi criado
display(f'Arquivo:{arquivo_excel} criado com sucesso!')

# 2. Ler o arquivo Excel e analisar os dados com NumPy
df = pd.read_excel(arquivo_excel)

# Pega apenas a coluna de vendas e converte para uma lista numérica
vendas = df['Vendas (R$)'].values

# Calcula estatísticas usando NumPy
media = np.mean(vendas) #média de valores
mediana = np.median(vendas) #mediana (valor central)
desvio = np.std(vendas) #desvio padrão(variação de dados)
total = np.sum(vendas) #soma total das vendas

# Mostra os resultados no terminal
display('Análise de Vendas da Empresa')
display(f'Total Anual: R${total:.2f}')
display(f'Média mensal: R${media:.2f}')
display(f'Mediana: R${mediana:.2f}')
display(f'Desvio padrão: R${desvio:.2f}')

# 3. Criar um gráfico com os dados de vendas

# Cria o gráfico de linha com os dados de vendas por mês
# Cria uma nova figura (janela de gráfico) 
# e define o tamanho: 10 de largura x 6 de altura (em polegadas)
plt.figure(figsize=(10,6))


# Cria o gráfico de linha:
# df['Mês']:    valores do eixo X (os meses do ano)
# vendas:       valores do eixo Y (os valores de vendas)
# marker='o':   adiciona um marcador (bolinha) em cada ponto do gráfico
# linestyle='-': usa linha contínua conectando os pontos
# color='green': define a cor da linha como verde
plt.plot(df['Mês'], vendas,marker='o',linestyle='-',color='green')

# título que aparece acima do gráfico
plt.title('Vendas mensais da empresa')

# nome do eixo X (horizontal)
plt.xlabel('Mês')

# nome do eixo Y (vertical)
plt.ylabel('Vendas (R$)')

# Gira os nomes dos meses em 45 graus
# para facilitar a leitura (evita que fiquem sobrepostos)
plt.xticks(rotation=45)


# Ativa uma grade de fundo no gráfico 
# (linhas claras para facilitar a visualização dos dados)
plt.grid(True)


# Ajusta automaticamente os elementos do gráfico 
# para que tudo caiba na imagem (sem cortar textos)
plt.tight_layout()


# Mostra o gráfico na tela 
# (abre uma janela com o gráfico ou exibe no notebook)
plt.show()


```
