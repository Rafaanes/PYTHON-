```python
'''
Enunciado:

A equipe de planejamento estratgico de uma empresa 
do setor automotivo est avaliando oportunidades 
de mercado com base em dados histricos de veculos. 
A anlise tem como objetivo identificar padres e 
relaes entre caractersticas tcnicas dos automveis 
e seus respectivos preos de mercado 
(MSRP - Manufacturer's Suggested Retail Price).

Com base em uma base de dados contendo informaes 
como marca, modelo, tipo de combustvel, potncia 
do motor e preo sugerido, solicita-se:

Elaborar uma anlise exploratria que permita 
visualizar e interpretar a variao mdia de 
preos dos veculos conforme o tipo de combustvel utilizado.

Essa anlise deve apoiar a tomada de deciso quanto:

    ao posicionamento de produtos,

     precificao por categoria,

    e  avaliao do impacto do tipo de combustvel 
    sobre o valor percebido pelo mercado.

A apresentao do resultado dever incluir 
um grfico claro e objetivo, bem como uma 
breve interpretao dos dados obtidos.

OBJETIVO:
=========
    Mdia de preo (MSRP) por tipo de combustvel.

    
ATENO: no terminal, copie e cole cada linha:
==================================
pip install pandas
pip install matplotlib

'''

# Importa a biblioteca pandas com o apelido "pd"
# Usada para manipular e analisar dados em tabelas (como planilhas)
import pandas as pd

# Importa a biblioteca matplotlib.pyplot com o apelido "plt"
# Usada para criar grficos em Python
import matplotlib.pyplot as plt

# Caminho do arquivo CSV contendo os dados dos veculos
# 02_Veiculos.csv
arquivo_csv = r'C:\Rafalla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\02_Veiculos.csv'

# L o arquivo CSV e carrega em uma tabela chamada "df" (DataFrame)
# O argumento delimiter=',' indica que os campos esto separados por vrgulas
df = pd.read_csv(arquivo_csv, sep = ',')

# Seleciona apenas as colunas "Engine Fuel Type" (tipo de combustvel) e "MSRP" (preo do veculo)
# dropna() remove linhas que tenham valores vazios (ausentes)
# groupby() agrupa os dados por tipo de combustvel
# mean() calcula a mdia de preo para cada tipo
# sort_values() ordena os resultados do mais caro para o mais barato (descending = False)
media_preco_por_combustivel = df[['Engine Fuel Type','MSRP']].dropna().groupby('Engine Fuel Type').mean('MSRP').sort_values(by='MSRP',ascending=False) #maior para o menor

# Cria uma nova figura para o grfico com tamanho 10 x 6 polegadas
plt.figure(figsize=(10,6))

# Cria um grfico de barras usando os dados calculados
# kind='bar' indica o tipo de grfico
# legend=False oculta a legenda
# color='green' define a cor das barras
media_preco_por_combustivel.plot(kind='bar', legend=False, color='green')

# Define o ttulo do grfico
plt.title('Mdia de preo (MSRP) por tipo de combustvel')

# Define o nome do eixo X (horizontal)
plt.xlabel('Tipo de Combustvel')

# Define o nome do eixo Y (vertical)
plt.ylabel('Preo Mdio (MSRP)')

# Adiciona linhas horizontais no fundo do grfico para facilitar a leitura
plt.grid(axis='y')

# Gira os nomes das categorias do eixo X em 45 graus para no ficarem sobrepostos
#ha = horizontal aligment (alinhamento horizontal)
plt.xticks(rotation=45, ha='right')

# Ajusta automaticamente o layout do grfico para evitar cortes
plt.tight_layout

# Exibe o grfico na tela
plt.show()






```
