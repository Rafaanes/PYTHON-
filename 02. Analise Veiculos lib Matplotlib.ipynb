```python
'''
Enunciado:

A equipe de planejamento estratégico de uma empresa 
do setor automotivo está avaliando oportunidades 
de mercado com base em dados históricos de veículos. 
A análise tem como objetivo identificar padrões e 
relações entre características técnicas dos automóveis 
e seus respectivos preços de mercado 
(MSRP - Manufacturer's Suggested Retail Price).

Com base em uma base de dados contendo informações 
como marca, modelo, tipo de combustível, potência 
do motor e preço sugerido, solicita-se:

Elaborar uma análise exploratória que permita 
visualizar e interpretar a variação média de 
preços dos veículos conforme o tipo de combustível utilizado.

Essa análise deve apoiar a tomada de decisão quanto:

    ao posicionamento de produtos,

    à precificação por categoria,

    e à avaliação do impacto do tipo de combustível 
    sobre o valor percebido pelo mercado.

A apresentação do resultado deverá incluir 
um gráfico claro e objetivo, bem como uma 
breve interpretação dos dados obtidos.

OBJETIVO:
=========
    Média de preço (MSRP) por tipo de combustível.

    
ATENÇÃO: no terminal, copie e cole cada linha:
==================================
pip install pandas
pip install matplotlib

'''

# Importa a biblioteca pandas com o apelido "pd"
# Usada para manipular e analisar dados em tabelas (como planilhas)
import pandas as pd

# Importa a biblioteca matplotlib.pyplot com o apelido "plt"
# Usada para criar gráficos em Python
import matplotlib.pyplot as plt

# Caminho do arquivo CSV contendo os dados dos veículos
# 02_Veiculos.csv
arquivo_csv = r'C:\Rafaélla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\02_Veiculos.csv'

# Lê o arquivo CSV e carrega em uma tabela chamada "df" (DataFrame)
# O argumento delimiter=',' indica que os campos estão separados por vírgulas
df = pd.read_csv(arquivo_csv, sep = ',')

# Seleciona apenas as colunas "Engine Fuel Type" (tipo de combustível) e "MSRP" (preço do veículo)
# dropna() remove linhas que tenham valores vazios (ausentes)
# groupby() agrupa os dados por tipo de combustível
# mean() calcula a média de preço para cada tipo
# sort_values() ordena os resultados do mais caro para o mais barato (descending = False)
media_preco_por_combustivel = df[['Engine Fuel Type','MSRP']].dropna().groupby('Engine Fuel Type').mean('MSRP').sort_values(by='MSRP',ascending=False) #maior para o menor

# Cria uma nova figura para o gráfico com tamanho 10 x 6 polegadas
plt.figure(figsize=(10,6))

# Cria um gráfico de barras usando os dados calculados
# kind='bar' indica o tipo de gráfico
# legend=False oculta a legenda
# color='green' define a cor das barras
media_preco_por_combustivel.plot(kind='bar', legend=False, color='green')

# Define o título do gráfico
plt.title('Média de preço (MSRP) por tipo de combustível')

# Define o nome do eixo X (horizontal)
plt.xlabel('Tipo de Combustível')

# Define o nome do eixo Y (vertical)
plt.ylabel('Preço Médio (MSRP)')

# Adiciona linhas horizontais no fundo do gráfico para facilitar a leitura
plt.grid(axis='y')

# Gira os nomes das categorias do eixo X em 45 graus para não ficarem sobrepostos
#ha = horizontal aligment (alinhamento horizontal)
plt.xticks(rotation=45, ha='right')

# Ajusta automaticamente o layout do gráfico para evitar cortes
plt.tight_layout

# Exibe o gráfico na tela
plt.show()






```
