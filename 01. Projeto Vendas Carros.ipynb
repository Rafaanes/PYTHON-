```python
''' 
Enunciado:

Crie um programa para ler os dados de vendas de carros de 
uma planilha Excel, calcule o preço médio de venda 
para cada marca de carro e apresenta os resultados. 

Mostre no console qual marca tem o preço médio mais 
alto e qual tem o preço médio mais baixo. 

O programa deverá gerar um gráfico de barras que 
ilustra visualmente o preço médio de venda de
cada marca, facilitando a comparação e a tomada de decisão.

'''

# Importa as bibliotecas para trabalhar com dados e gráficos
import pandas as pd
import matplotlib.pyplot as plt

# Coloque o caminho do arquivo Excel no seu computador aqui:
caminho_arquivo = r'C:\Rafaélla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Projeto\Bases Projeto\01.Vendas Carros.xlsx'

# Lê os dados da planilha chamada 'Planilha1'
df = pd.read_excel(caminho_arquivo,sheet_name = 'Planilha1')

# Calcula o preço médio de venda para cada marca
preco_medio = df.groupby ('Marca') ['Preço Venda'].mean()

# Organiza os preços médios do maior para o menor
preco_medio = preco_medio.sort_values(ascending=False)

# Mostra no console qual marca tem o preço médio mais alto
display('Marca com preço médio mais alto: ', preco_medio.idxmax())

#display('Marca com preço médio mais alto: ', round preco_medio.max(),2))

# Mostra no console qual marca tem o preço médio mais baixo
display('Preço médio mais baixo: ', round(preco_medio.min(),2))


# Cria um gráfico de barras para mostrar o preço médio de cada marca
#display (preco_medio[0])


# Coloca o título do gráfico


# Nomeia os eixos


# Gira os nomes das marcas para facilitar a leitura


# Adiciona linhas horizontais para facilitar a comparação


# Ajusta o gráfico para não cortar nada


# Mostra o gráfico na tela
```
