```python
'''
CENÁRIO: Análise da Distribuição de Lucro por Pedido
====================================================

Você recebeu uma base de dados contendo os pedidos realizados por 
diferentes lojas de uma rede varejista. Essa base contém informações 
como o valor de venda, o custo e o identificador da loja e do produto.

O objetivo deste exercício é calcular o lucro por pedido, 
analisar a distribuição desses lucros e criar uma visualização 
gráfica clara e profissional que permita interpretar 
rapidamente o comportamento dos dados.

Tarefas:
========
    Carregar o arquivo pedidos.csv contendo os dados dos pedidos.

    Criar uma nova coluna chamada 'Lucro', calculando a 
    diferença entre Receita_Venda e Custo_Venda.

    Exibir o lucro total da empresa com base nos dados.

    Construir um gráfico de histograma com curva de densidade 
    para representar a distribuição dos lucros por pedido.

    Permitir que o usuário personalize a cor das colunas 
    do gráfico, alterando uma variável chamada cor_colunas.

Requisitos técnicos:
===================
Utilizar as bibliotecas pandas, matplotlib e seaborn.

No terminal:
==========================

pip install pandas
pip install matplotlib
pip install seaborn

'''


# ============================================
# 1. Importação das bibliotecas
# ============================================

# pandas é usado para ler, tratar e analisar os dados em tabelas (DataFrames)
import pandas as pd

# matplotlib.pyplot é usado para criar gráficos simples como barras, linhas, etc.
import matplotlib.pyplot as plt

# seaborn é uma biblioteca de gráficos mais avançada, que facilita a visualização de dados
import seaborn as sns

# Define o estilo visual dos gráficos (fundo branco com grade e cores suaves)
sns.set_theme(style = 'whitegrid', palette='pastel')

# ============================================
# 2. Carregamento do arquivo CSV
# ============================================

# Caminho do arquivo CSV com os dados de vendas
# 03_Pedidos.csv
arquivo_csv = r'C:\Rafaélla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\03_Pedidos.csv'

# Leitura do arquivo CSV com separador vírgula (',') e carregamento no DataFrame 'df'
df = pd.read_csv(arquivo_csv, sep = ',', encoding = 'latin-1')

# Exibe as 5 primeiras linhas da tabela para dar uma olhada nos dados
df.head()

# ============================================
# 3. Cálculo do Lucro
# ============================================

# Cria uma nova coluna chamada 'Lucro'
# O lucro é calculado subtraindo o custo da receita de venda (receita - custo)
df['Lucro'] = df['Receita_Venda'] - df['Custo_Venda']

# Soma todos os valores da coluna 'Lucro' para saber o lucro total
lucro_total = df['Lucro'].sum()

# Mostra o resultado na tela com separador de milhar e duas casas decimais
display(f'Lucro total: R$ {lucro_total:,.2f}')

# ============================================
# 4. Personalização da cor do gráfico
# ============================================

# Aqui você pode trocar a cor das colunas do gráfico conforme preferir
# Exemplos de cores: 'crimson', 'royalblue', 'skyblue', 'orange', '#FF5733'
cor_colunas = 'royalblue'

# ============================================
# 5. Criação do gráfico de distribuição de lucro
# ============================================

# Define o tamanho da figura do gráfico (largura=10, altura=5)
plt.figure(figsize = (10,5))

# Cria um histograma (gráfico de colunas) para mostrar como os lucros estão distribuídos
# bins=30 → define o número de colunas no gráfico (quanto maior, mais detalhado)
# kde=True → adiciona uma linha suavizada (curva) sobre o gráfico para facilitar a leitura
# color → define a cor das colunas (definida acima)
sns.histplot(df['Lucro'], bins=30, kde=True, color=cor_colunas)

# Define o título do gráfico
plt.title('Distribuição do Lucro por Pedido')

# Define o texto do eixo X
plt.xlabel ('Lucro por Pedido')

# Define o texto do eixo Y
plt.ylabel('Frequência')

# Mostra o gráfico na tela
plt.show()



```
