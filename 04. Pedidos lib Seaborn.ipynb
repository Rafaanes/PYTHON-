```python
'''
CENRIO: Anlise da Distribuio de Lucro por Pedido
====================================================

Voc recebeu uma base de dados contendo os pedidos realizados por 
diferentes lojas de uma rede varejista. Essa base contm informaes 
como o valor de venda, o custo e o identificador da loja e do produto.

O objetivo deste exerccio  calcular o lucro por pedido, 
analisar a distribuio desses lucros e criar uma visualizao 
grfica clara e profissional que permita interpretar 
rapidamente o comportamento dos dados.

Tarefas:
========
    Carregar o arquivo pedidos.csv contendo os dados dos pedidos.

    Criar uma nova coluna chamada 'Lucro', calculando a 
    diferena entre Receita_Venda e Custo_Venda.

    Exibir o lucro total da empresa com base nos dados.

    Construir um grfico de histograma com curva de densidade 
    para representar a distribuio dos lucros por pedido.

    Permitir que o usurio personalize a cor das colunas 
    do grfico, alterando uma varivel chamada cor_colunas.

Requisitos tcnicos:
===================
Utilizar as bibliotecas pandas, matplotlib e seaborn.

No terminal:
==========================

pip install pandas
pip install matplotlib
pip install seaborn

'''


# ============================================
# 1. Importao das bibliotecas
# ============================================

# pandas  usado para ler, tratar e analisar os dados em tabelas (DataFrames)
import pandas as pd

# matplotlib.pyplot  usado para criar grficos simples como barras, linhas, etc.
import matplotlib.pyplot as plt

# seaborn  uma biblioteca de grficos mais avanada, que facilita a visualizao de dados
import seaborn as sns

# Define o estilo visual dos grficos (fundo branco com grade e cores suaves)
sns.set_theme(style = 'whitegrid', palette='pastel')

# ============================================
# 2. Carregamento do arquivo CSV
# ============================================

# Caminho do arquivo CSV com os dados de vendas
# 03_Pedidos.csv
arquivo_csv = r'C:\Rafalla Sena - Python\Python 2\Python 2 - Clarify\Aluno\Aula_02\Aula\Bases\03_Pedidos.csv'

# Leitura do arquivo CSV com separador vrgula (',') e carregamento no DataFrame 'df'
df = pd.read_csv(arquivo_csv, sep = ',', encoding = 'latin-1')

# Exibe as 5 primeiras linhas da tabela para dar uma olhada nos dados
df.head()

# ============================================
# 3. Clculo do Lucro
# ============================================

# Cria uma nova coluna chamada 'Lucro'
# O lucro  calculado subtraindo o custo da receita de venda (receita - custo)
df['Lucro'] = df['Receita_Venda'] - df['Custo_Venda']

# Soma todos os valores da coluna 'Lucro' para saber o lucro total
lucro_total = df['Lucro'].sum()

# Mostra o resultado na tela com separador de milhar e duas casas decimais
display(f'Lucro total: R$ {lucro_total:,.2f}')

# ============================================
# 4. Personalizao da cor do grfico
# ============================================

# Aqui voc pode trocar a cor das colunas do grfico conforme preferir
# Exemplos de cores: 'crimson', 'royalblue', 'skyblue', 'orange', '#FF5733'
cor_colunas = 'royalblue'

# ============================================
# 5. Criao do grfico de distribuio de lucro
# ============================================

# Define o tamanho da figura do grfico (largura=10, altura=5)
plt.figure(figsize = (10,5))

# Cria um histograma (grfico de colunas) para mostrar como os lucros esto distribudos
# bins=30  define o nmero de colunas no grfico (quanto maior, mais detalhado)
# kde=True  adiciona uma linha suavizada (curva) sobre o grfico para facilitar a leitura
# color  define a cor das colunas (definida acima)
sns.histplot(df['Lucro'], bins=30, kde=True, color=cor_colunas)

# Define o ttulo do grfico
plt.title('Distribuio do Lucro por Pedido')

# Define o texto do eixo X
plt.xlabel ('Lucro por Pedido')

# Define o texto do eixo Y
plt.ylabel('Frequncia')

# Mostra o grfico na tela
plt.show()



```
