Como foi a evolução das exportações para a frança ao longo dos anos?

Crie uma tabela com os anos:

    2019,

e o valor US$ de exportações

1) faça um agrupamento por ano e some "US$ FOB"
2) formate os valores da coluna "US$ FOB" em US$
3) exibir os resultados anuais totais

```python
def formatar(valor):
    valor_formatado = f' US$ {valor:,}'
    return valor_formatado

faturamento_por_plano = dados.groupby('Year').sum('US$ FOB')
faturamento_por_plano = faturamento_por_plano[['US$ FOB']]

faturamento_por_plano['US$ FOB'] = faturamento_por_plano ['US$ FOB'].apply(formatar)

display(faturamento_por_plano)
```
